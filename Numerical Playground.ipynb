```python
# Daniel Boe
# September 7, 2019
# Numicial solution to transport equation

'''The goal of this script is to illustrate how python can be used to 
numerically solve Partial Differential Equations'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline


'''Define the Initial Condition (a normal distribution in this case)'''
def ini_condition(x):
    sigma=0.5
    xBar=5

    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-xBar)**2/(2*sigma*2))


step_size=0.1
a=1
time_step=step_size/a/2
SpaceDomain=np.arange(0,15,step_size)
TimeDomain=np.arange(0,5,time_step)

DF=pd.DataFrame({'x':SpaceDomain})
DF[0]=ini_condition(DF.x)
DF.set_index('x',inplace=True) 

for t in TimeDomain[1:]:
    DF[round(t,5)]=a*time_step/step_size*(DF.iloc[:,-1].shift().fillna(0)\
                - DF.iloc[:,-1])+DF.iloc[:,-1]


# Generate the Mesh
t,x=np.meshgrid(TimeDomain,SpaceDomain)

fig=plt.figure(dpi=120)
ax=fig.gca(projection='3d')
ax.plot_wireframe(x,t,DF,color='slategray')
fig.suptitle('Pure Advection: '+r'$\frac{\partial u}{\partial t} + a\frac{\partial u}{\partial x}=0$',weight='bold')

plt.show()
```

```python
# Daniel Boe
# September 8, 2019
# Numicial solution to transport equation with Diffusion

'''The goal of this script is to illustrate how python can be used to 
numerically solve Partial Differential Equations'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

def ini_condition(x):
    sigma=0.5
    xBar=5

    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-xBar)**2/(2*sigma*2))

# Define the constants
u=1
D=0.75

# Set the step size
step_size=0.1
dt=0.01/2

# Construct the domain
SpaceDomain=np.arange(0,15,step_size)
TimeDomain=np.arange(0,5,dt)

DF=pd.DataFrame({'x':SpaceDomain})
DF[0]=ini_condition(DF.x)
DF.set_index('x',inplace=True) 

for t in TimeDomain[1:]:
    DF[round(t,5)]=dt*(D/step_size**2*DF.iloc[:,-1].shift(-1).fillna(0) - \
          1/step_size**2*((2*D+u*step_size)*DF.iloc[:,-1]-\
          (D+u*step_size)*DF.iloc[:,-1].shift().fillna(0))) +DF.iloc[:,-1]


#Generate the Mesh
t,x=np.meshgrid(TimeDomain,SpaceDomain)

fig=plt.figure(dpi=120)
ax=fig.gca(projection='3d')
ax.plot_wireframe(x,t,DF,color='steelblue')
fig.suptitle('Advection and Diffusion: '+r'$\frac{\partial u}{\partial t}'+
             r'+ a\frac{\partial u}{\partial x} - D\frac{\partial^2 u}{\partial x^2}=0$'
             ,weight='bold')
plt.show()
```
