```python
import bokeh.plotting as blt
import bokeh.io as io
import bokeh.models as mdl
import pandas as pd 
import numpy as np
```

```python
DF=pd.read_csv('EnergyData.csv')
```

```python
DF=DF.filter(['Start','Active Calories (kcal)'])
DF.rename(columns={'Start':'Date','Active Calories (kcal)':'Energy'},inplace=True)
```

```python
DF.Date=pd.to_datetime(DF.Date)
DF.set_index('Date',inplace=True)
DF.Energy=DF.Energy*4.184/1000 # Convert to MJ
```

```python
io.output_notebook()
```

```python
TOOLS = "crosshair,pan,wheel_zoom,box_zoom,reset,box_select,lasso_select"

fig=blt.figure(title='Daily Energy Data',x_axis_label='Date',y_axis_label='Energy [MJ]',x_axis_type='datetime',
              tools=TOOLS,plot_width=800, plot_height=500)

#fig.circle(DF.index,DF.Energy,color='seagreen',size=15,fill_alpha=0.5,line_color=None)
fig.line(DF.index,DF.Energy,color='seagreen',line_width=3,alpha=0.5)


blt.show(fig)
```

```python
DF['Month']=DF.index.month
Aggregate = DF.groupby(by='Month').mean()


TOOLS = "crosshair,pan,wheel_zoom,box_zoom,reset,box_select,lasso_select"

fig2=blt.figure(title='Daily Energy Data',x_axis_label='Month',y_axis_label='Energy [MJ]',
              tools=TOOLS,plot_width=800, plot_height=500)

#fig2.circle(DF.index,DF.Energy,color='seagreen',size=15,fill_alpha=0.5,line_color=None)
fig2.line(Aggregate.index,Aggregate.Energy,color='seagreen',line_width=3,alpha=0.5)


blt.show(fig2)
```
